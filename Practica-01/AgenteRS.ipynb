````python
REGLAS = {
    "sin-pelota": "pedir-pelota",
    "pedir-pelota": "pedir-carga",
    "pedir-carga": "pelota-lista",
    "pelota-lista": "pide-lanza",
    "pide-lanza": "pelota-lanzada",
    "pelota-lanzada": "espera",
}


class AgenteRS:
    def __init__(self, reglas: dict[str, str]):
        self.reglas = reglas

    def actuar(self, percepcion, accion_basica=""):
        if not percepcion:
            return accion_basica

        return self.reglas.get(percepcion, accion_basica)


try:
    agente = AgenteRS(REGLAS)
    while percepcion := input("Indicar percepcion: "):
        accion = agente.actuar(percepcion, "Error")
        print(f"[Agente]==> {accion}")
except (KeyboardInterrupt, EOFError):
    print("\b\b  \nSaliendo del programa...")
````
