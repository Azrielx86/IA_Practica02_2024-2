````python
REGLAS: dict[str, str] = {
    "sin-pelota": "pedir-pelota",
    "con-pelota": "pedir-carga",
    "pelota-lista": "pide-lanza",
    "pelota-lanzada": "espera",
}


MODELO: dict[tuple[str, str, str], str] = {
    ("sin-pelota", "pedir-pelota", "pelota"): "con-pelota",
    ("con-pelota", "pedir-carga", "carga"): "pelota-lista",
    ("pelota-lista", "pide-lanza", "lanza"): "pelota-lanzada",
    ("pelota-lanzada", "espera", "recibe"): "sin-pelota",
}


class Modelos:
    def __init__(self, modelo: dict, reglas: dict, estado_ini="", accion_ini="") -> None:
        self.modelo = modelo
        self.reglas = reglas
        self.estado_ini = estado_ini
        self.estado = self.estado_ini
        self.accion_ini = accion_ini
        self.accion = None
        self.u_accion = self.accion_ini

    def actuar(self, percepcion: str) -> str:
        if not percepcion:
            return self.accion_ini

        clave = (self.estado, self.u_accion, percepcion)

        self.estado = self.modelo.get(clave)
        accion = self.reglas.get(self.estado, "")

        if not accion:
            self.estado = self.estado_ini
            self.accion = self.accion_ini
            self.u_accion = self.accion_ini
            return self.accion_ini

        self.u_accion = accion
        return accion


try:
    agente = Modelos(MODELO, REGLAS, "sin-pelota", "pedir-pelota")
    while percepcion := input("Indica una percepción: "):
        accion = agente.actuar(percepcion)
        print(f"[Agente]==> {accion}")
except (KeyboardInterrupt, EOFError):
    print("\b\b  \nSaliendo del programa...")
````
