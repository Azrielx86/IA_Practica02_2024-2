````python
ACCIONES = {
    "pelota": "almacena",
    "pelota,carga": "prepara",
    "pelota,carga,lanza": "lanza",
    "pelota,carga,lanza,recibe": "espera",
}


class AgenteTabla:
    def __init__(self, acciones: dict[str, str]):
        self.acciones = acciones
        self.percepciones = ""

    def actuar(self, percepcion, accion_basica=""):
        if not percepcion:
            return accion_basica

        self.percepciones += ("," if self.percepciones else "") + percepcion

        accion = self.acciones.get(self.percepciones, accion_basica)
        if accion is accion_basica:
            self.percepciones = ""

        return accion


try:
    agente = AgenteTabla(ACCIONES)
    while percepcion := input("Indicar percepción: "):
        accion = agente.actuar(percepcion, "espera")
        print(f"[Agente]==> {accion}")
except (KeyboardInterrupt, EOFError):
    print("\b\b  \nSaliendo del programa...")
````
